```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path ="D:/atchaya/main.py/Telco-Customer-Churn.csv"  # Using forward slashes

df = pd.read_csv(file_path)

# Standardize column names
df.columns = df.columns.str.strip().str.lower()

# Convert 'churn' column to numerical (assuming 'Yes' = 1, 'No' = 0)
if 'churn' in df.columns:
    df['churn'] = df['churn'].map({'Yes': 1, 'No': 0})

# Select numerical columns for correlation analysis
numeric_cols = df.select_dtypes(include=['number'])

# Compute correlation matrix
correlation_matrix = numeric_cols.corr()

# Sort features by correlation with 'churn'
if 'churn' in correlation_matrix.columns:
    churn_correlation = correlation_matrix['churn'].drop('churn').sort_values(ascending=False)
else:
    print("âŒ 'Churn' column not found in dataset. Check column names:", df.columns)
    exit()

# Display the top features correlated with churn
print("ðŸ“Š Top Features Most Correlated with Customer Churn:")
print(churn_correlation.head(5))  # Show top 5 correlated features

# Visualization: Correlation Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

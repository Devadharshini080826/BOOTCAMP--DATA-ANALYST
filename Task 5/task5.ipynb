```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data (Income, Life Expectancy, and Population for Different Countries)
data = {
    "Year": [2000, 2000, 2000, 2010, 2010, 2010, 2020, 2020, 2020],
    "Country": ["USA", "India", "China", "USA", "India", "China", "USA", "India", "China"],
    "Income": [40000, 5000, 10000, 45000, 7000, 12000, 50000, 9000, 15000],
    "Life Expectancy": [77, 63, 70, 78, 66, 74, 79, 69, 76],
    "Population": [300, 1000, 1300, 320, 1200, 1350, 330, 1400, 1400]  # Bubble size
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Function to plot Bubble Chart for a given year
def plot_bubble_chart(year):
    df_year = df[df["Year"] == year]

    plt.figure(figsize=(10, 6))
    plt.scatter(df_year["Income"], df_year["Life Expectancy"], 
                s=df_year["Population"], alpha=0.5, edgecolors="k")

    # Add country labels
    for i, row in df_year.iterrows():
        plt.text(row["Income"], row["Life Expectancy"], row["Country"], fontsize=10)

    plt.xlabel("Income (GDP per Capita)")
    plt.ylabel("Life Expectancy (Years)")
    plt.title(f"Life Expectancy vs Income ({year})")
    plt.xscale("log")  # Log scale for better visualization
    plt.grid(True, linestyle="--", alpha=0.7)
    
    plt.show()

# Example: Plot for Year 2000
plot_bubble_chart(2000)
```

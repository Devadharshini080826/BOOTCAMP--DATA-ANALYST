```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data (Year vs. Average Income)
data = {
    "Year": [2000, 2005, 2010, 2015, 2020, 2025],
    "Average Income (USD)": [25000, 30000, 35000, 42000, 50000, 58000]
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Create a Line Chart
plt.figure(figsize=(8, 6))
plt.plot(df["Average Income (USD)"], df["Year"], marker="o", linestyle="-", color="blue", label="Income Trend")

# Set Titles and Labels
plt.title("Change in Average Income Over the Years", fontsize=14)
plt.xlabel("Average Income (USD)", fontsize=12)
plt.ylabel("Year", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```

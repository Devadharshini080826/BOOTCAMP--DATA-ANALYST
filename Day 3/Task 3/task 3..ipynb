```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data (Happiness Score over Years for Different Countries)
data = {
    "Year": [2015, 2016, 2017, 2018, 2019, 2020],
    "USA": [7.2, 7.1, 7.0, 6.9, 6.8, 6.7],
    "India": [5.5, 5.4, 5.3, 5.2, 5.0, 4.8],
    "Germany": [7.0, 7.2, 7.4, 7.5, 7.6, 7.7],
    "Japan": [5.9, 5.8, 5.7, 5.6, 5.5, 5.4]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the Line Chart
plt.figure(figsize=(8, 6))

# Loop through each country and plot
for country in df.columns[1:]:  # Skip 'Year' column
    plt.plot(df[country], df["Year"], marker="o", linestyle="-", label=country)

# Set Titles and Labels
plt.title("Happiness Score Changes Over the Years", fontsize=14)
plt.xlabel("Happiness Score (WHR)", fontsize=12)
plt.ylabel("Year", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```

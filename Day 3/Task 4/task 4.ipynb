```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data (Year vs. Number of Cell Phones per Person)
data = {
    "Year": [1990, 1995, 2000, 2005, 2010, 2015, 2020, 2025],
    "Cell Phones per Person": [0.02, 0.1, 0.3, 0.7, 1.1, 1.5, 2.0, 2.5]  # Hypothetical growth data
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Create a Line Chart
plt.figure(figsize=(8, 6))
plt.plot(df["Cell Phones per Person"], df["Year"], marker="o", linestyle="-", color="blue", label="Cell Phones Trend")

# Set Titles and Labels
plt.title("Rise of Cell Phones per Person Over Time", fontsize=14)
plt.xlabel("Number of Cell Phones per Person", fontsize=12)
plt.ylabel("Year", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data (Year vs. Average Life Expectancy)
years = [1960, 1970, 1980, 1990, 2000, 2010, 2020]
life_expectancy = [52.5, 58.3, 63.2, 67.1, 70.5, 72.8, 74.6]

# Set style
sns.set_style("whitegrid")

# Create the figure and axis
plt.figure(figsize=(8, 5))

# Scatter plot
plt.scatter(life_expectancy, years, color='blue', label="Life Expectancy")

# Line plot for trend
plt.plot(life_expectancy, years, linestyle="dashed", color="red", alpha=0.6)

# Labels and Title
plt.xlabel("Average Life Expectancy (Years)")
plt.ylabel("Year")
plt.title("Life Expectancy Over the Years")

# Show the plot
plt.legend()
plt.show()
```
